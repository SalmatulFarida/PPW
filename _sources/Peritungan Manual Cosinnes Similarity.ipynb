# Perhitungan Manual Cosinnes Similarity

```python
import pandas as pd
import chardet
import requests

# URL file CSV di GitHub
url = 'https://raw.githubusercontent.com/SalmatulFarida/PPW/gh-pages/P%20Manual%20Cosinnes%20Similarity.csv'

# Deteksi encoding
with requests.get(url, stream=True) as r:
    result = chardet.detect(r.content)
    encoding = result['encoding']

# Membaca file CSV dengan encoding yang berbeda-beda
encodings_to_try = ['utf-8', 'latin-1', 'ISO-8859-1']
df = None

for encoding_try in encodings_to_try:
    try:
        df = pd.read_csv(url, delimiter=";", encoding=encoding_try)
        encoding = encoding_try  # Menggunakan encoding yang berhasil
        break  # Keluar dari loop jika berhasil dibaca
    except UnicodeDecodeError:
        print(f"Failed to decode using {encoding_try}")

if df is None:
    raise ValueError("Failed to read the CSV file with any of the tried encodings.")

# Mengisi nilai yang kosong
df_filled = df.fillna('')

# Menampilkan data frame dan encoding yang digunakan
print("Data Frame:")
print(df_filled)
print("\nEncoding Used:", encoding)
```
