# Crawling Pta

Tugas :

Lakukan crawling data pada pta.trunojoyo.ac.id

data disimpan dalam bentuk .csv

data yang diambil : judul, nama penulis, pembimbing I, pembimbing II, abstrak

Library :

- import requests : import requests adalah pernyataan dalam bahasa pemrograman Python yang digunakan untuk mengimpor modul atau pustaka bernama "requests". Modul ini memberikan fungsionalitas untuk membuat permintaan HTTP dalam program Python, sehingga memungkinkan program berkomunikasi dengan server web atau API untuk mengambil atau mengirim data.
- import csv : adalah pernyataan dalam bahasa pemrograman Python yang digunakan untuk mengimpor modul CSV. Modul ini menyediakan fungsi-fungsi untuk membaca dan menulis file CSV, yang merupakan format umum untuk menyimpan data terstruktur dalam bentuk teks.
- import pandas : adalah pernyataan dalam bahasa pemrograman Python yang digunakan untuk mengimpor modul pandas. Pandas adalah pustaka open-source yang menyediakan struktur data dan alat analisis data untuk Python. Pustaka ini sangat populer dalam analisis data, manipulasi data, dan pembersihan data.
- import warnings : warnings adalah modul bawaan dalam Python yang memungkinkan Anda mengelola peringatan (warnings) yang muncul selama eksekusi program.


```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
headers = {
        'Access-Control-Allow-Origin': '*',
        'Access-Control-Allow-Methods': 'GET',
        'Access-Control-Allow-Headers': 'Content-Type',
        'accept': '/',
        'accept-encoding': 'gzip, deflate',
        'accept-language': 'en,mr;q=0.9',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

```python
list_judul = []
list_penulis = []
list_pembimbing1 = []
list_pembimbing2 = []
list_abstak = []
url = f"https://pta.trunojoyo.ac.id/c_search/byprod/9"
req = requests.get(url,headers=headers,timeout=5,verify=False)
for i in range(2, 144):
    soup = BeautifulSoup(req.content, 'html.parser')
    results = soup.find(id="content_journal")
    elements = results.find_all("li", attrs={"data-id":"id-1"})
#     print(elements)
    for element in elements:
        info = element.find_all("div")[0]
        info_nama = info.find_all("div")
#         print(info_nama)
        link_abstrak = element.find_all("div")[-1].find("a")["href"]
#         ekstrak info
        judul = info.find("a", class_="title").text
        penulis, pembimbing1, pembimbing2 = [nama.text.split(":")[1].strip() for nama in info_nama]

#         ekstrak abstrak

#         print(link_abstrak)
        req = requests.get(link_abstrak,headers=headers,timeout=5,verify=False)
        soup_abstrak =  BeautifulSoup(req.content, 'html.parser')
        results_abstrak = soup_abstrak.find_all("p")

        abstrak = results_abstrak[0].text

        list_judul.append(judul)
        list_penulis.append(penulis)
        list_pembimbing1.append(pembimbing1)
        list_pembimbing2.append(pembimbing2)
        list_abstak.append(abstrak)

    url = f"https://pta.trunojoyo.ac.id/c_search/byprod/9{i}"
    req = requests.get(url,headers=headers,timeout=50,verify=False)
    if(req.status_code == 404):
        break
```

```python
df = pd.DataFrame({
    "Judul" : list_judul,
    "Penulis" : list_penulis,
    "Pembimbing I" : list_pembimbing1,
    "Pembimbing II" : list_pembimbing2,
    "Abstrak" : list_abstak
})
df
#df.to_csv("ptaUtm.csv", index=True)
```
